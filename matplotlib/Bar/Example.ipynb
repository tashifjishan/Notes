```python
import matplotlib.pyplot as plt

# Data
products = ["Laptop", "Mobile", "Tablet", "Watch"]
sales = [120, 180, 90, 140]
sd = [20, 10, 25, 30]

# Figure
plt.figure(figsize=(9, 5.5))

# Bar chart with error bars
bars = plt.bar(
    products,
    sales,
    yerr=sd,                  # Error bars
    capsize=10,                # Cap size
    color=["#4C78A8", "#F58518", "#54A24B", "#E45756"],
    edgecolor="black",
    linewidth=1.2,
    alpha=0.9,
    width=0.65,
    error_kw={
        "elinewidth": 1,
        "ecolor": "black",
        "capthick": 1
    },
    label="Average Sales"
)

# Value labels
print(bars)
print(sd)
print(zip(bars, sd))
for bar, err in zip(bars, sd):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + err + 3,
        f"{bar.get_height()}",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

# Labels and title
plt.title("Product Sales with Standard Deviation", fontsize=16, weight="bold")
plt.xlabel("Products", fontsize=12)
plt.ylabel("Units Sold", fontsize=12)

# Grid
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.gca().set_axisbelow(True)

# Remove unnecessary borders
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.legend()
plt.tight_layout()
plt.show()
```
